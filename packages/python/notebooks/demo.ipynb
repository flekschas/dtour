```python
from arro3.io import read_parquet
import polars as pl

df = pl.from_arrow(read_parquet("../data/mair-2022-tumor-006-ozette.pq"))
df.head(10)
```

```python
cols = [col for col in df.columns if col.endswith("Windsorized")]
cols
```

```python
from sklearn.preprocessing import StandardScaler

scaled_df = StandardScaler().fit_transform(df.select(cols).to_numpy())
```

```python
from umap import UMAP

umap = UMAP(n_components=4)
embedding = umap.fit_transform(scaled_df)
```

```python
result = pl.DataFrame(
    {f"umap_{i}": embedding[:, i] for i in range(embedding.shape[1])}
).with_columns(df.get_column("faustLabels").alias("label"))

result.write_parquet("../data/mair-2022-tumor-006-ozette-umap-4d.pq")
```
